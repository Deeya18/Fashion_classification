```python
'''
class_labels:
  0 -> T-Shirt/top
  1 -> Trouser
  2 -> Pullover
  3 -> Dress
  4 -> Coat
  5 -> Sandal
  6 -> Shirt
  7 -> Sneaker
  8 -> Bag
  9 -> Ankle boot

classify the given  input from above class using neural network
image shape 28 x 28 pixel (Gray scale)

'''
```

```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
class_labels = ["T-Shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]



plt.imshow(x_train[0], cmap = "Greys")
```

```python
plt.figure(figsize = (16,16))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_train[i], cmap = "Greys")
  plt.axis('off')
  plt.title(class_labels[y_train[i]] + "=" + str(y_train[i]), fontsize = 20)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = [28,28]),
    keras.layers.Dense(100, activation = "relu"),
    keras.layers.Dense(10, activation = "softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train, epochs = 1)
```

```python
model.fit(x_train, y_train, epochs = 10)
```

```python
model.evaluate(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
y_pred[0].round(2)
```

```python
np.argmax(y_pred[0].round(2))
```

```python
y_test[0]
```

```python
plt.figure(figsize = (16,16))
for i in range(25):
  plt.subplot(5,5,i+1)

  plt.imshow(x_test[i], cmap = "Greys")
  plt.axis('off')
  plt.title("Actual = {} \n predicted = {}".format(class_labels[y_test[i]] , class_labels[np.argmax(y_pred[i])]) )
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,[ np.argmax(i) for i in y_pred])
plt. figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = "d", cmap = "Blues")
```
